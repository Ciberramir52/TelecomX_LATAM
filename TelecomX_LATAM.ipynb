#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd

df = pd.read_json('https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json')
df
```

#ðŸ”§ TransformaciÃ³n

```python
pd.set_option('display.max_columns', None)
```

```python
customer = df['customer'].apply(pd.Series)
df = df.drop(columns=['customer']).join(customer)
df
```

```python
phone = df['phone'].apply(pd.Series)
df = df.drop(columns=['phone']).join(phone)
df
```

```python
internet = df['internet'].apply(pd.Series)
df = df.drop(columns=['internet']).join(internet)
df
```

```python
account = df['account'].apply(pd.Series)
df = df.drop(columns=['account']).join(account)
df
```

```python
charges = df['Charges'].apply(pd.Series)
df = df.drop(columns=['Charges']).join(charges)
df
```

```python
df.info()
```

```python
df["Total"] = df["Total"].apply(lambda x: 0 if x.strip() == "" else x)
```

```python
df["Total"] = df["Total"].astype("float64")
df.info()
```

```python
mask = df["customerID"].duplicated(keep=False)
df[mask]
```

```python
df = df[df['Churn'].apply(lambda x: x.strip().lower()).isin(['yes', 'no'])]
df
```

```python
df['cuentas_diarias'] = df['Monthly'] / 30
df
```

```python
cols_to_replace = ['Churn', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling']
df[cols_to_replace] = df[cols_to_replace].replace({'Yes': 1, 'No': 0, 'No phone service': 0, 'No internet service': 0})
df
```

```python
df = df.rename(columns={
    "customerID": "id_cliente",
    "Churn": "abandono",
    "gender": "genero",
    "SeniorCitizen": "es_adulto_mayor",
    "Partner": "tiene_pareja",
    "Dependents": "tiene_dependientes",
    "tenure": "antiguedad_meses",
    "PhoneService": "tiene_servicio_telefonico",
    "MultipleLines": "tiene_lineas_multiples",
    "InternetService": "tipo_servicio_internet",
    "OnlineSecurity": "seguridad_en_linea",
    "OnlineBackup": "respaldo_en_linea",
    "DeviceProtection": "proteccion_dispositivo",
    "TechSupport": "soporte_tecnico",
    "StreamingTV": "tiene_streaming_tv",
    "StreamingMovies": "tiene_streaming_peliculas",
    "Contract": "tipo_contrato",
    "PaperlessBilling": "tiene_facturacion_electronica",
    "PaymentMethod": "metodo_pago",
    "Monthly": "cargos_mensuales",
    "Total": "cargos_totales",
})
df
```

#ðŸ“Š Carga y anÃ¡lisis

```python
df[['antiguedad_meses', 'cargos_mensuales', 'cargos_totales', 'cuentas_diarias']].describe()
```

```python
frec_abandono = df['abandono'].value_counts()
frec_abandono_prc = df['abandono'].value_counts(normalize=True) * 100
frec_abandono
frec_abandono_prc
```

```python
import matplotlib.pyplot as plt
```

```python
colors = plt.cm.Set1(range(len(frec_abandono_prc)))
ax = frec_abandono_prc.plot(kind='bar', color=colors)
ax.set_ylabel('Porcentaje (%)')
ax.set_xlabel('Abandono')
ax.set_title('Porcentaje de clientes que abandonan')
ax.set_xticklabels(['No abandonan', 'Abandonan'], rotation=0)
# plt.xticks(rotation=0)
plt.tight_layout()
```

```python
tab = (pd.crosstab(df["genero"], df["abandono"], normalize="index") * 100)[1]
colors = plt.cm.Set1(range(len(tab)))

ax = tab.plot(kind="bar", color=colors)
ax.set_ylabel("Porcentaje (%)")
ax.set_xlabel("GÃ©nero")
ax.set_title("DistribuciÃ³n de abandono por gÃ©nero")
plt.xticks(rotation=0)
plt.tight_layout()
```

```python
tab = (pd.crosstab(df["antiguedad_meses"], df["abandono"], normalize="index") * 100)
quit_sorted = tab.sort_values(by=1, ascending=False).head(10)[1]

ax = quit_sorted.plot(kind="bar", color='green')
ax.set_ylabel("Porcentaje (%)")
ax.set_xlabel("Antiguedad Meses")
ax.set_title("DistribuciÃ³n de abandono por Antiguedad")
plt.xticks(rotation=0)
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
df_churn = df[df['abandono'] == 1]
sns.histplot(data=df_churn, x="cargos_totales",
             multiple="dodge", alpha=0.7, bins=30, color='red', edgecolor='black')
plt.title("Distribucion de Evasiones por Costo Total")
plt.xlabel("Costo Total")
plt.ylabel("Numero de clientes")
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data=df_churn, x="cargos_mensuales",
             multiple="dodge", alpha=0.7, bins=30, color='#F44336', edgecolor='black')
plt.title("Distribucion de Evasiones por Costo Mensual")
plt.xlabel("Costo Mensual")
plt.ylabel("Numero de clientes")
plt.show()
```
